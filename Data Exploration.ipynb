Import vessel information, using `nmsi` as unique identifier.

```python
vessels = pd.read_csv("fonnesbeck_20130313/vessel_information.csv", index_col='mmsi')
vessels.head()
```

Filter out vessels with more than one name (unique identity):

```python
vessels = vessels[vessels.num_names==1]
```

Merge all towing ships:

```python
towing_types = ['Tug', 'Towing', 'BigTow', 'Towing/Tug', 'BigTow/Towing']
vessels.type[[v in towing_types for v in vessels.type]] = 'All Tow'
```

Merge cargo types:

```python
vessels.type[[v in ['Cargo/Other', 'Cargo/Unknown', 'Cargo/Reserved'] for v in vessels.type]] = 'Cargo'
```

Filter out unwanted vessel types:

```python
valid_types = [
 'Cargo',
 'Fishing',
 'Tanker',
 'Passenger',
 'All Tow',
 'HSC',
 'Other',
 'Pleasure',
 'Sailing',
 'WIG']

vessels = vessels[[v in valid_types for v in vessels.type]]
```

Plot segment speeds in active (red) and inactive (blue) SMA's:

```python
def plot_segment_speeds(sma=None, vtype=None):
    data = segments_merged
    if sma is not None:
        data = data[data.sma==sma]
    if vtype is not None:
        data = data[data.type==vtype]
    time = data.st_time
    speed = data.avg_sog
    
    figure(figsize=(16,6))
    plot_date(dates.date2num(time[data.active==0]), speed[segments_merged.active==0], 'b.', alpha=0.3)
    plot_date(dates.date2num(time[data.active==1]), speed[segments_merged.active==1], 'r.', alpha=0.3)
    ylabel('Average Speed')
    if sma is not None:
        title('Segment speeds in {0} for {1}'.format(sma,vtype))
```

```python
for sma in transit_segment.sma.unique():
    plot_segment_speeds(sma)
```

Segment speeds by vessel type within a single SMA:

```python
sma = 'NY'
for vtype in segments_merged.type.unique():
    plot_segment_speeds(sma, vtype)
```
