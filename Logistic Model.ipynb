```python
date = datetime.date
sma_dates = {'NC_GA': [(date(y,11,1), date(y+1,4,30)) for y in range(2008,2012)],
            'NOR': [(date(y,11,1), date(y+1,4,30)) for y in range(2008,2012)],
            'SE': [(date(y,11,15), date(y+1,4,15)) for y in range(2008,2012)],
            'MOR': [(date(y,11,1), date(y+1,4,30)) for y in range(2008,2012)],
            'PHI': [(date(y,11,1), date(y+1,4,30)) for y in range(2008,2012)],
            'NY': [(date(y,11,1), date(y+1,4,30)) for y in range(2008,2012)],
            'CCB': [(date(y,1,1), date(y,5,15)) for y in range(2009,2012)],
            'ORP': [(date(y,3,1), date(y,4,15)) for y in range(2009,2012)],
            'BI': [(date(y,11,1), date(y+1,4,30)) for y in range(2008,2012)],
            'GSC': [(date(y,4,1), date(y,7,31)) for y in range(2009,2012)]
            }
```

```python
def find_season(rec):
    sma = rec['sma']
    stime = rec['st_time']
    d = stime.date()
    seasons = sma_dates[sma]
    for i,daterange in enumerate(seasons):
        if daterange[0] <= d < daterange[1]:
            return i
    return NaN
```

```python
dataset['season'] = dataset.apply(find_season, axis=1)
```

## Logistic regression model

One approach for detecting behavioral differences in ships over time is to compare vessels that have transited the SMA in multiple seasons. Here is a simple logistic regression that compares ships that appeared in both the first and last season of a particular SMA. The response variable is the event that the PDGT10 value in the later season is less than or equal to the PGDT10 value in the early season. We expect this to be a function of the initial PDGT10 value, with a greater potential for a response in vessels that are initally fast. 

The model is:

$$\text{logit}[Pr(x_{late} < x_{early})] = \alpha + \beta x_{early}$$

A sample run of the model is below. The plot at the bottom shows the result for cargo vessels in `NC_GA`; the data points as blue dots, and the lines are samples from the posterior distribution for the parameters, with each line essentially a plausible "true" relationship. You can see greater uncertainty at lower initial PGDT10, where data are low and more certainty at higher values of initial PGDT10. 

```python
def generate_diffs(sma, contrast=[0,-1], dataset=dataset, min_size=20):
    ship_means = []
    # Loop over ship types
    for i,ship in enumerate(valid_types):
        data_subset = dataset[(dataset.type==ship) & (dataset.sma==sma) & (dataset.season.notnull())]
        seasons = np.sort(data_subset.season.unique())
        if len(seasons) > 1:
            means = [data_subset[['mmsi','pdgt10']][data_subset.season==d].groupby('mmsi').mean() for d in seasons[contrast]]
            means_merged = means[0].merge(means[1], suffixes=['_early','_late'], how='inner', right_index=True, left_index=True)
            if len(means_merged)>=min_size:
                means_merged['type'] = ship
                ship_means.append(means_merged)
    if ship_means:
        return pd.concat(ship_means)
```

```python
sma_list = dataset.sma.unique()
sma_list
```

```python
invlogit = lambda x: np.exp(x)/(1 + np.exp(x)) 
```

```python
iterations = 20000
```

```python
scenario_traces = []

for sma in dataset.sma.unique():

    sma_data = generate_diffs(sma)
    
    if sma_data is not None:
    
        X = pd.get_dummies(sma_data.type).values
        early, late = sma_data[[u'pdgt10_early', u'pdgt10_late']].T.values
        y = (late <= early).astype(int)
        
        k = min(X.shape[1], valid_types)
        
        with Model() as model:
        
            alpha = Normal('alpha', 0, 0.001, shape=k)
        
            beta = Normal('beta', 0, 0.001, shape=k)
        
            p = tinvlogit(t.dot(X, alpha) + t.dot(X, beta)*early)
        
            slower = Bernoulli('slower', p=p, observed=y)
            
            start = model.test_point
            step = NUTS()
            
            tr = psample(iterations, step, start=start) 
                
            scenario_traces.append(tr)
        
    else:
        
        scenario_traces.append(None)
        
```

```python
def plot_logistic(ship, sma, trace, dataset=dataset, season_a=0, season_b=-1, nlines=100, burnin=5000, min_size=20):

    if trace is not None:
        try:
            trace = trace.traces[0]
        except AttributeError:
            pass
        
        # Create subset based on conditions
        data_subset = dataset[(dataset.type==ship) & (dataset.sma==sma) & (dataset.season.notnull())]
        # Data by season
        season_data = [data_subset[['mmsi','pdgt10']][data_subset.season==d].groupby('mmsi').mean() 
                       for d in np.sort(data_subset.season.unique())]
        
        merged_seasons = season_data[season_a].merge(season_data[season_b], suffixes=['_early','_late'], how='inner', 
                                right_index=True, left_index=True)
        if len(merged_seasons) < min_size:
            return
        x = merged_seasons.pdgt10_early
        y = merged_seasons.pdgt10_late<=merged_seasons.pdgt10_early
        plt.figure()
        scatter(x.values, y.values + np.random.randn(len(y))*0.01, color='k', marker='.')
        title('{0} in {1}'.format(ship, sma))
        
        # Extract alpha and beta samples from trace
        sma_data = generate_diffs(sma)
        ship_index = pd.get_dummies(sma_data.type).columns.tolist().index(ship)
        a = trace[burnin:].samples['alpha'].value[:,ship_index]
        b = trace[burnin:].samples['beta'].value[:,ship_index]
    
        # Plot lines
        for i in range(nlines):
            xvals = np.linspace(0, 100)
            plot(xvals, invlogit(a[-(i+1)] + b[-(i+1)]*xvals), 'r-', alpha=0.05)
        xlim(-1,101)
        ylim(-.2, 1.2)
        yticks([0,1], [0,1])
        ylabel('P(late < early)')
        xlabel('PDGT10 early')
```

```python
from pymc import forestplot, traceplot

for tr in scenario_traces: 
    if tr is not None:
        plt.figure()
        traceplot(tr[iterations/2:])
```

```python
for sma in sma_list:
    plot_logistic('Cargo', sma, scenario_traces[sma_list.tolist().index(sma)], burnin=iterations/2)
```

```python
for sma in sma_list:
    plot_logistic('Tanker', sma, scenario_traces[sma_list.tolist().index(sma)], burnin=iterations/2)
```

Here are the estimates of the effect of early PDGT10 on the probability that the late PDGT10 is less than that during the early period. Logistic regression parameters are a little tricky to interpret because they are parameters of a non-linear function, but we are generally interested in large positive values, since they correspond to a higher probability of a reduced PDGT10 after the intervention when the pre-intervention PDGT10 is higher.

```python
from pymc import summary

for tr in scenario_traces:
    if tr is not None:
        summary(tr[iterations/2:], vars=['beta'])
```

### Simulate random PGDT10 data

```python
from numpy.random import beta

early_sim = beta(1, 1, size=500)*100
late_sim = beta(1, 1, size=500)*100
y = late_sim < early_sim
```

```python
with Model() as sim_model:

    alpha = Normal('alpha', 0, 0.001)

    beta = Normal('beta', 0, 0.001)

    p = tinvlogit(alpha + beta*early_sim)

    slower = Bernoulli('slower', p=p, observed=y)
    
    start = sim_model.test_point
    step = NUTS()
    
    sim_tr = sample(10000, step, start=start) 
        
```

### NOVA program

To look at the effect of the first set of NOVAs, which were sent during the 2010/2011 season, we will compare PDGT10 in the 2009/2010 season to the 2011/2012 season.

The logisitic relationships map PDGT10 values pre-NOVA to the probability that the post-NOVA PDGT10 will be lower that before the intervention. If the intervention is *ineffective* we would expect that there would be no relationship; ships should on average be just as fast after post-NOVA than pre-NOVA, so we would predict a line hovering around 0.5. 

If the intervention is effective, we would expect logistic relationships that bend upwards. This would reflect a higher probability of a lower post-NOVA PDGT10 for ships that are moving fast before the intervention, but a lower probability for ships that are already relatively compliant.
