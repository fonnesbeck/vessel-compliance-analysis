```python
def find_season(rec):
    sma = rec['sma']
    stime = rec['st_time']
    d = stime.date()
    seasons = sma_dates[sma]
    for i,daterange in enumerate(seasons):
        if daterange[0] <= d < daterange[1]:
            return i
    return NaN
```

## Logistic regression model

One approach for detecting behavioral differences in ships over time is to compare vessels that have transited the SMA in multiple seasons. Here is a simple logistic regression that compares ships that appeared in both the first alnd last season of a particular SMA. The response variable is the event that the PDGT10 value in the later season is less than or equal to the PGDT10 value in the early season. We expect this to be a function of the initial PDGT10 value, with a greater potential for a response in vessels that are initally fast. 

The model is:

$$\text{logit}[Pr(x_{late} < x_{early})] = \alpha + \beta x_{early}$$

A sample run of the model is below. The plot at the bottom shows the result for cargo vessels in `NC_GA`; the data points as blue dots, and the lines are samples from the posterior distribution for the parameters, with each line essentially a plausible "true" relationship. You can see greater uncertainty at lower initial PGDT10, where data are low and more certainty at higher values of initial PGDT10. 

```python
# Choose the most prevalent ship types
stypes = 4
ships = valid_types[:stypes]
ships.sort()
```

```python
def generate_diffs(sma, cutoff=stypes, contrast=[0,-1], dataset=dataset):
    ship_means = []
    # Loop over ship types
    for i,ship in enumerate(valid_types[:stypes]):
        data_subset = dataset[(dataset.type==ship) & (dataset.sma==sma) & (dataset.season.notnull())]
        seasons = np.sort(data_subset.season.unique())
        means = [data_subset[['mmsi','pdgt10']][data_subset.season==d].groupby('mmsi').mean() for d in seasons[contrast]]
        means_merged = means[0].merge(means[1], suffixes=['_early','_late'], how='inner', right_index=True, left_index=True)
        means_merged['type'] = ship
        ship_means.append(means_merged)
    return pd.concat(ship_means)
```

```python
sma_list = dataset.sma.unique()
sma_list
```

```python
from pymc import *

def run_model(sma, ship):
    
    sma_data = generate_diffs(sma)
    
    sma_data = sma_data[sma_data.type==ship]
    early, late = sma_data[[u'pdgt10_early', u'pdgt10_late']].T.values
    y = (late <= early).astype(int)
    
    alpha = Normal('alpha', 0, 0.001, value=0)
    
    beta = Normal('beta', 0, 0.001, value=0)

    p = Lambda('p', lambda a=alpha, b=beta: invlogit(a + b*early))

    slower = Bernoulli('slower', p=p, value=y, observed=True)
    
    return locals()
```

```python
from pymc import *
import theano.tensor as t

def tinvlogit(x):
    return t.exp(x)/(1 + t.exp(x)) 

def run_model(sma, ship):
    
    sma_data = generate_diffs(sma)
    
    sma_data = sma_data[sma_data.type==ship]
    early, late = sma_data[[u'pdgt10_early', u'pdgt10_late']].T.values
    y = (late <= early).astype(int)
        
    with Model() as model:
    
        alpha = Normal('alpha', 0, 0.001)
    
        beta = Normal('beta', 0, 0.001)
    
        p = tinvlogit(alpha + beta*early)
    
        slower = Bernoulli('slower', p=p, observed=y)
        
        start = model.test_point
        #h = approx_hess(start)
        #step = NUTS(model.vars, h)
        step = Metropolis(model.vars)
        
        tr = sample(10000, step, start=start) 
            
    return tr
```

```python
invlogit = lambda x: np.exp(x)/(1 + np.exp(x)) 
```

```python
def plot_logistic(ship, sma, trace, dataset=dataset, season_a=0, season_b=-1, nlines=200):
    
    plt.figure()
    # Create subset based on conditions
    data_subset = dataset[(dataset.type==ship) & (dataset.sma==sma) & (dataset.season.notnull())]
    # Data by season
    season_data = [data_subset[['mmsi','pdgt10']][data_subset.season==d].groupby('mmsi').mean() 
                   for d in np.sort(data_subset.season.unique())]
    
    merged_seasons = season_data[season_a].merge(season_data[season_b], suffixes=['_early','_late'], how='inner', 
                            right_index=True, left_index=True)
    x = merged_seasons.pdgt10_early
    y = merged_seasons.pdgt10_late<=merged_seasons.pdgt10_early
    scatter(x.values, y.values)
    title('{0} in {1}'.format(ship, sma))
    
    # Extract alpha and beta samples from trace
    sma_data = generate_diffs(sma)
    ship_index = pd.get_dummies(sma_data.type).columns.tolist().index(ship)
    a = trace.samples['alpha'].value[:,ship_index]
    b = trace.samples['beta'].value[:,ship_index]

    # Plot lines
    for i in range(nlines):
        xvals = np.linspace(0, 100)
        plot(xvals, invlogit(a[-(i+1)] + b[-(i+1)]*xvals), 'm-', alpha=0.05)
    xlim(-1,101)
    ylim(-.2, 1.2)
    yticks([0,1], [0,1])
    ylabel('P(late < early)')
    xlabel('PDGT10 early')
```

## Select comparisons

### NOVA program

To look at the effect of the first set of NOVAs, which were sent during the 2010/2011 season, we will compare PDGT10 in the 2009/2010 season to the 2011/2012 season.

The logisitic relationships map PDGT10 values pre-NOVA to the probability that the post-NOVA PDGT10 will be lower that before the intervention. If the intervention is *ineffective* we would expect that there would be no relationship; ships should on average be just as fast after post-NOVA than pre-NOVA, so we would predict a line hovering around 0.5. 

If the intervention is effective, we would expect logistic relationships that bend upwards. This would reflect a higher probability of a lower post-NOVA PDGT10 for ships that are moving fast before the intervention, but a lower probability for ships that are already relatively compliant.

```python
# Before and after Nov-Dec 2010
nova_comparisons = dict(BI=(1,3), 
                        CCB=(1,2), 
                        GSC=(1,2),
                        MOR=(1,3),
                        NC_GA=(1,3),
                        NOR=(1,3),
                        NY=(1,3),
                        ORP=(1,2),
                        PHI=(1,3),
                        SE=(1,3))
```

```python
nova_traces = []

for sma in dataset.sma.unique():
    
    sma_data = generate_diffs(sma, nova_comparisons[sma])
    
    X = pd.get_dummies(sma_data.type).values
    early, late = sma_data[[u'pdgt10_early', u'pdgt10_late']].T.values
    y = (late <= early).astype(int)
    
    k = min(X.shape[1], stypes)
    
    with Model() as model:
    
        alpha = Normal('alpha', 0, 0.001, shape=k)
    
        beta = Normal('beta', 0, 0.001, shape=k)
    
        p = tinvlogit(dot(X, alpha) + dot(X, beta)*early)
    
        slower = Bernoulli('slower', p=p, observed=y)
        
        start = model.test_point
        #h = approx_hess(start)
        #step = NUTS(model.vars, h)
        step = Metropolis(model.vars)
        
        tr = sample(20000, step, start=start) 
            
    nova_traces.append(tr)
```

Here are the estimates of the effect of pre-NOVA PDGT10 on the probability that the post-NOVA PDGT10 is less than that during pre-NOVA. Logistic regression parameters are a little tricky to interpret because they are parameters of a non-linear function, but we are generally interested in large positive values, since they correspond to a higher probability of a reduced PDGT10 after the intervention when the pre-intervention PDGT10 is higher. Parameters whose credible (HPD) interval do not contain zero would roughly correspond to "significant" parameters in the classical hypothesis-testing approach.

```python
for sma in sma_list:
    sma_data = generate_diffs(sma, nova_comparisons[sma])
    ships = pd.get_dummies(sma_data.type).columns
    sma_trace = nova_traces[sma_list.tolist().index(sma)]
    print('Parameters for {0}: {1}'.format(sma, ships.tolist()))
    foo = summary(sma_trace, vars=[beta])
```
